```python
import nltk
from nltk.corpus import stopwords
import re
import glob

f = lambda x: x.replace("\\", "/")
text_files = glob.glob('*.txt')
text_files = (list(map(f, text_files)))
print(text_files)
results = []
my_data = []



for file in text_files:
    doc = open(file, "r", encoding= "utf8")
    for line in doc:
        text= str(line)
        #print(text)
        text = ' '.join([word for word in text.split() if word not in (stopwords.words('english'))])
        text = re.sub(r'^https?:\/\/.*[\r\n]*', '', text, flags=re.MULTILINE)
        text = re.sub(r'((1-\d{3}-\d{3}-\d{4})|(\(\d{3}\) \d{3}-\d{4})|(\d{3}-\d{3}-\d{4}))$', '', text)
        text = re.sub(r'[^\w\s]','',text)
        text = text.replace(" ", ",")
        text = str(text)
        my_data.append(text)
        words = [x for x in my_data if x]
        words = ''.join(words)
    results.append(words)
    print(file)
    my_data.clear()
```

```python
import pandas as pd
percentile_list = pd.DataFrame(
    {'file_name': text_files,
     'words': results
    })


df = percentile_list
df['i'] = df['words'].astype(str).map(len)

df.update(df[['file_name']].applymap('"{}"'.format))
df['result'] = df[['file_name', 'words']].agg(':'.join, axis=1)
df = df[['result']]
df.to_csv('result.dat')
df
```
